# NLP Assignment: Tokenization, Stemming, and Lemmatization

This notebook demonstrates various NLP preprocessing techniques using NLTK.

```

import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Sample Text

```

text = "NLTK's tokenizers are amazing! I'm learning NLP, and it's fun 😄 #AI #NLP"
print(text)
```

## Tokenization

```

from nltk.tokenize import WhitespaceTokenizer, wordpunct_tokenize, TreebankWordTokenizer, TweetTokenizer, MWETokenizer

wt = WhitespaceTokenizer()
print("Whitespace Tokenization:", wt.tokenize(text))

print("Punctuation-based Tokenization:", wordpunct_tokenize(text))

treebank = TreebankWordTokenizer()
print("Treebank Tokenization:", treebank.tokenize(text))

tweet = TweetTokenizer()
print("Tweet Tokenization:", tweet.tokenize(text))

mwe = MWETokenizer([('machine', 'learning'), ('natural', 'language', 'processing')])
sample_mwe_text = "I love machine learning and natural language processing"
print("MWE Tokenization:", mwe.tokenize(sample_mwe_text.split()))
```

## Stemming

```

from nltk.stem import PorterStemmer, SnowballStemmer

porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in treebank.tokenize(text)]
print("Porter Stemming:", porter_stems)

snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in treebank.tokenize(text)]
print("Snowball Stemming:", snowball_stems)
```

## Lemmatization

```

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word) for word in treebank.tokenize(text)]
print("Lemmatization:", lemmas)
```
