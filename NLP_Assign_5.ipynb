# NLP Assignment: Semantic Relationships using WordNet

This notebook demonstrates how to identify synonymy, antonymy, and hypernymy using WordNet in NLTK.

```

import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Sample Words

```

words = ["good", "bad", "car", "dog"]
print(words)
```

## Synonymy (Synonyms)

```

from nltk.corpus import wordnet as wn

def get_synonyms(word):
    synonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return synonyms

for word in words:
    print(f"Synonyms of '{word}':", get_synonyms(word))
```

## Antonymy (Antonyms)

```

def get_antonyms(word):
    antonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())
    return antonyms

for word in words:
    print(f"Antonyms of '{word}':", get_antonyms(word))
```

## Hypernymy (Hypernyms)

```

def get_hypernyms(word):
    hypernyms = set()
    for syn in wn.synsets(word):
        for hyper in syn.hypernyms():
            hypernyms.add(hyper.name())
    return hypernyms

for word in words:
    print(f"Hypernyms of '{word}':", get_hypernyms(word))
```

## Conclusion

WordNet helps identify semantic relationships between words, enabling better understanding of meaning in NLP applications.
