```python
import nltk
from collections import defaultdict, Counter
import random
```

```python
# Sample training text
text = """
Artificial intelligence is transforming the world.
Artificial intelligence helps machines learn from data.
Machine learning is a part of artificial intelligence.
AI systems can understand language and predict text.
"""
```

```python
# Preprocess text
tokens = text.lower().split()

# Choose N (for example trigram)
N = 3
```

```python
# Build N-gram model
ngrams = defaultdict(Counter)

for i in range(len(tokens) - N + 1):
    context = tuple(tokens[i:i+N-1])   # previous N-1 words
    next_word = tokens[i+N-1]          # predicted word
    ngrams[context][next_word] += 1
```

```python
# Autocomplete function
def autocomplete(input_text, n=3):
    words = input_text.lower().split()

    if len(words) < N-1:
        return "Enter more words"

    context = tuple(words[-(N-1):])

    if context in ngrams:
        suggestions = ngrams[context].most_common(n)
        return [word for word, count in suggestions]
    else:
        return "No prediction found"
```

```python
# Example
print(autocomplete("artificial intelligence"))
```
