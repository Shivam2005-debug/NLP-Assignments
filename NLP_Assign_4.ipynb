# NLP Assignment 4: Named Entity Recognition (NER)

This notebook demonstrates a **Named Entity Recognition (NER)** system using **NLTK** and evaluates it using **Accuracy, Precision, Recall, and F1-score**.

**Dataset:** Custom real-world sentences (news & social media style).

## Step 1: Import Libraries

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker_tab')
nltk.download('maxent_ne_chunker')
nltk.download('words')

from nltk import word_tokenize, pos_tag, ne_chunk
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## Step 2: Entity Extraction Example

```python
text = "Barack Obama was born in Hawaii and served as President of the United States."
tokens = word_tokenize(text)
pos = pos_tag(tokens)
ner_tree = ne_chunk(pos)
print(ner_tree)
```

## Step 3: Custom Evaluation Dataset

```python
dataset = [
    ("Apple was founded by Steve Jobs", ['ENT','O','O','ENT','ENT']),
    ("Google is headquartered in California", ['ENT','O','O','O','O','ENT']),
    ("Elon Musk leads Tesla", ['ENT','ENT','O','ENT']),
    ("Narendra Modi is Prime Minister of India", ['ENT','ENT','O','O','ENT','ENT','ENT'])
]

def get_predictions(sentence):
    tokens = word_tokenize(sentence)
    pos = pos_tag(tokens)
    tree = ne_chunk(pos)
    labels = []
    for item in tree:
        if hasattr(item, 'label'):
            for _ in item:
                labels.append('ENT')
        else:
            labels.append('O')
    return labels

y_true, y_pred = [], []

for sent, true_labels in dataset:
    pred_labels = get_predictions(sent)
    y_true.extend(true_labels)
    y_pred.extend(pred_labels)
```

## Step 4: Evaluation Metrics

```python
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label='ENT')
recall = recall_score(y_true, y_pred, pos_label='ENT')
f1 = f1_score(y_true, y_pred, pos_label='ENT')

print(f"Accuracy : {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall   : {recall:.2f}")
print(f"F1-Score : {f1:.2f}")
```

## Conclusion

The NER system effectively extracts named entities from real-world text. Using a custom dataset avoids dependency errors while still allowing proper evaluation using standard NLP metrics.
